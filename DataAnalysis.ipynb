### Missing Values

```python
## Here we will not handle the missing values that we do in feature engineering
## Here we will check the percentage of nan values present in each Feature
## step - 1 :- make the list of features which has missing values
feature_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
## step -2 :- pirnt the feature name and the percentage of missing values

for feature in feature_with_na:
    print(feature , np.round(dataset[feature].isnull().mean(),4), '% missing values')
```

## Since they are many missing values, we need to find the relationship between missing values and Sales Price
Plot diagram for this relationship

```python
for feature in feature_with_na:
    data = dataset.copy()
    
    ## Let's make a variable that indicates 1 if observation is missing or 0 if the observation is not missing
    data[feature] =np.where(data[feature].isnull(),1,0)
    
    ## Let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

Here with the relation between the missing values and the dependent variable is clearly visible. So We need to replace these nan values with something meaningful which we will do in the Feature Engineering section.

#### Numerical Variables

```python
##List of numerical values
numerical_feature = [feature for feature in dataset.columns if dataset[feature].dtype != "O"]

print("Number of Numerical Feature {}".format(len(numerical_feature)))

dataset[numerical_feature].head()
```

##### Temporal Variable (Eg. Datetime Variables)
From the Dataset we have 4 year Variables. We have to extract infromation from the datetime varibales like no of years or no of days

```python
# List of variables that contain year information
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
## Let's  explore the content of these year information
for feature in year_feature:
    print(feature , dataset[feature].unique())
```

```python
dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("Year Sold")
plt.ylabel("Median House Price")
plt.title("House Price vs YearSold")
```

```python
## Compare the difference between all years feature with SalePrice
for feature in year_feature:
    if feature!= 'YrSold':
        data=dataset.copy()
        ## We will capture the difference between year variable and year the house sold
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
        
```

##### Discrete Feature

```python
##numerical variables are usually of 2 types
#Continous variable and Discrete variable
discrete_feature = [feature for feature in numerical_feature if len(dataset[feature].unique()) < 25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
#Find the relationship between them and Sale Price
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar(color=colors)
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

##### Continous Feature

```python
Continous_feature = [feature for feature in numerical_feature if  feature not in discrete_feature+year_feature+['Id']]
print("Continous Variables Count: {}".format(len(Continous_feature)))
```

```python
for feature in Continous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
##Using Logarithmic transformation
for feature in Continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data["SalePrice"]=np.log(data["SalePrice"])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(feature)
        plt.show()
        
```

#### Outliers

```python
for feature in Continous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
