# House Prices: Advanced Regression Techniques :- Feature Engineering (for test data) 

We will be performing all the below steps in Feature Engineering

1. Missing Values
2. Temporal variables
3. Categorical variables: remove rare labels
4. Standaise the values of the variables to the same range


### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

### Upload dataset in notebook

```python
dataset = pd.read_csv("test.csv")
```

```python
dataset.head()
```

## Missing Values

##### Categorical Features

```python
cat_feature = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=="O"]
```

```python
for feature in cat_feature:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
# replacing Missing values in categorical feature
def replace_cat_nan(dataset,cat_feature):
    data=dataset.copy()
    data[cat_feature]=data[cat_feature].fillna("Missing")
    return data
dataset = replace_cat_nan(dataset,cat_feature)
dataset[cat_feature].isnull().sum()
```

```python
dataset.head(15)
```

##### Numerical Feature

```python
num_feature = [feature for feature in dataset if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O' ]
for feature in num_feature:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replacing Numerical Features missing variables with median of that feature
for feature in num_feature:
    median_value  = dataset[feature].median()
    dataset[feature+"_nan"] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)

dataset[num_feature].isnull().sum()    
```

```python
dataset.head(50)
```

### Temporal Features

```python
for feature in ['YearBuilt',"YearRemodAdd","GarageYrBlt"]:
    dataset[feature] = dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

### Numerical Features (Log Transformation)

```python
num_feature =['LotFrontage','LotArea','1stFlrSF','GrLivArea']
for feature in num_feature:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset[num_feature].head()
```

## Handling Rare Categorical Feature
we will remove categorical variables that are present less than 1% of the observations

```python
cat_feature = [feature for feature in dataset.columns if dataset[feature].dtypes=='O']
cat_feature
```

```python
for feature in cat_feature:
    temp = dataset.groupby(feature).count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(100)
```

## Converting Categorical variables into Numerical Variables

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for feature in cat_feature:
    dataset[feature]=encoder.fit_transform(dataset[feature])
```

```python
dataset.head()
```

```python
feature1 = [feature for feature in dataset.columns if feature not in ["Id"]]
```

```python
feature1
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
for feature in feature1:
    dataset[feature] = minmax.fit_transform(dataset[[feature]])
dataset.head(100)
```

```python
dataset.to_csv('Y_test.csv',index = False)
```
